```python
import numpy as np
import matplotlib.pyplot as plt
```

##### FFT of a polynomial function

```python
SLOPE = 0
INTERCEPT = 5
T0 = 0
TF = 10
DT = 0.001
T0_signal = 3
TF_signal = 4
```

```python
time = np.arange(T0,TF,DT)
flat_function = time * SLOPE + INTERCEPT
square_function = flat_function * ((T0_signal < time) & (time < TF_signal))
```

```python
plt.plot(time, flat_function)
plt.plot(time, square_function)
plt.show()
```

```python
freqs = np.fft.fftfreq(n=time.size, d=DT)

fhat_flat = np.fft.fft(flat_function)
psd_flat = np.conjugate(fhat_flat) / time.size

fhat_square = np.fft.fft(square_function)
psd_square = np.conjugate(fhat_square) / time.size

plt.plot(freqs, psd_flat)
plt.plot(freqs, psd_square)
plt.xlim(-5,5)
plt.show()
```
